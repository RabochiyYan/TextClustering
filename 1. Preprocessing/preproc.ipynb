```python
def remove_non_cyrillic(text):

    if not isinstance(text, str):
        return ""

    cleaned_text = re.sub(r'[^–∞-—è–ê-–Ø—ë–Å\s]', '', text)
    return cleaned_text.lower()
```

```python
data_test_proc = data_test.apply(remove_non_cyrillic)

for i, t in enumerate(data_test_proc.values):
    print(f'{i} - {t}\n\n')
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
print(stopwords.words("russian"))
print(len(stopwords.words("russian")))
stopw = stopwords.words("russian")
```

```python
def stop_word_del(text):
    buf = []
    for word in text:
        if not(word in stopw):
            buf.append(word)
    return buf
```

```python
for i in range(5):
    print('–ü—Ä–æ–≤–µ—Ä–∫–∞ ', i+1)
    print(data5.values[i])
    print()
    print(stop_word_del(data5.values[i]))
    print('\n'*2)
```

### 1.5 –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏—é/—Å—Ç–µ–º–º–∏–Ω–≥, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–¥—Ä–∞–∑—É–º–µ–≤–∞–µ—Ç —Å–æ–∫—Ä–∞—â–µ–Ω–∏–µ –≤—Å–µ—Ö —Å–ª–æ–≤ —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —á—Ç–æ–±—ã –æ–Ω–∏ –≤—ã–≥–ª—è–¥–µ–ª–∏ –æ–¥–∏–Ω–∞–∫–æ–≤–æ, –Ω–æ –ø—Ä–∏ —ç—Ç–æ–º —Å–æ—Ö—Ä–∞–Ω—è–ª–∏ —Å–≤–æ—é —É–Ω–∏–∫–∞–ª—å–Ω–æ—Å—Ç—å –∑–∞ —Å—á—ë—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –±–∞–∑–æ–≤–æ–π –∏–ª–∏ –∫–æ—Ä–Ω–µ–≤–æ–π —Ñ–æ—Ä–º—ã –æ–¥–Ω–æ–≥–æ –∏ —Ç–æ–≥–æ –∂–µ —Å–ª–æ–≤–∞.

–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –≤—ã—Å—á–∏—Ç–∞–≤–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–ª–æ–≤ –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ. –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è, —á—Ç–æ –æ–Ω–æ –¥–æ–ª–∂–Ω–æ —É–º–µ–Ω—å—à–∏—Ç—å—Å—è, —Ç.–∫. —Ü–µ–ª—å —Å—Ç–µ–º–º–∏–Ω–≥–∞ –∏ –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏–∏ –≤ —É–Ω–∏—Ñ–∏–∫–∞—Ü–∏–∏ —Å–ª–æ–≤ —Å —Å—Ö–æ–∂–∏–º —Å–º—ã—Å–ª–æ–º

```python
def eval_unique(df):
    un_words = set()
    for text in df.values:
        un_words.update(text)
    return len(un_words)

print(f'–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–ª–æ–≤ –≤ –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ: {eval_unique(data6)}')
```

```python
from nltk.stem.snowball import SnowballStemmer
from nltk.tokenize import word_tokenize

stemmer = SnowballStemmer("russian")
text = "–õ–∏—Å—Ç–æ–≤—ã–µ –ª–∏—Å—Ç–æ—á–∫–∏ –ª–∏—Å—Ç –ª–∏—Å—Ç–≤–∞ –ª–∏—Å—Ç–≤–µ –ø–æ—á–µ–º—É —Ç–∞–∫"
tokens = word_tokenize(text)
stemmed_words = [stemmer.stem(word) for word in tokens]
print(stemmed_words)
```

#### pymorchpy3

```python
import pymorphy3
morph = pymorphy3.MorphAnalyzer()

text = "–õ–∏—Å—Ç–æ–≤—ã–µ –ª–∏—Å—Ç–æ—á–∫–∏ –ª–∏—Å—Ç –ª–∏—Å—Ç–≤–∞ –ª–∏—Å—Ç–≤–µ –ø–æ—á–µ–º—É —Ç–∞–∫"

def lemmatize(text):
    words = text.split() # —Ä–∞–∑–±–∏–≤–∞–µ–º —Ç–µ–∫—Å—Ç –Ω–∞ —Å–ª–æ–≤–∞
    res = list()
    for word in words:
        p = morph.parse(word)[0]
        res.append(p.normal_form)

    return res

print(lemmatize(text))
```

```python
morph = pymorphy3.MorphAnalyzer()

def lemmatize_pymorph(text):
    res = list()
    for word in text:
        p = morph.parse(word)[0]
        res.append(p.normal_form)

    return res

data_test_lemmatized1 = data_test.apply(lemmatize_pymorph)


print(f'–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–ª–æ–≤ –≤ –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ: {eval_unique(data6)}')
print(f'–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–ª–æ–≤ –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ –ø–æ—Å–ª–µ –ª–µ–º–º–∞—Ç–∏–∑–∞—Ü–∏–∏ —Å –ø–æ–º–æ—â—å—é pymorphy3: {eval_unique(data_test_lemmatized1)}')
```
